```python
import geopandas as gpd
```

```python
trees_data = gpd.read_file("SF Urban Tree Canopy.geojson")
```

```python
trees_data.head()
```

```python
import geoplot
import geoplot.crs as gcrs
subset = trees_data.sample(frac=0.03)
geoplot.polyplot(subset, projection=gcrs.AlbersEqualArea(), edgecolor='darkgreen', facecolor='lightgreen', linewidth=.3,
    figsize=(12, 8))
```

```python
geoplot.polyplot(trees_data, projection=gcrs.AlbersEqualArea(), edgecolor='darkgreen', facecolor='lightgreen', linewidth=.3,
    figsize=(25, 25))
```

```python
twenty_sample = trees_data.sample(frac=0.5)
geoplot.polyplot(twenty_sample, projection=gcrs.AlbersEqualArea(), edgecolor='darkgreen', facecolor='lightgreen', linewidth=.3,
    figsize=(25, 25))
```
